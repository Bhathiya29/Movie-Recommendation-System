### Calculate a weighted rating

```
# WR =(V + (v+m)) x R + (m+(v+m)) x C
```

```
# v - is the number of votes for a movie
# m - is the minimum votes required to be listed in the chart 
# R - is the average rating of the movie
# C - is the average rating across all movies
```

```
m = movies['vote_count'].quantile(0.9)
m
```

```
 C = movies ['vote_average'].mean()
 C
```

```
movies_filtered = movies.copy().loc[movies['vote_count'] >=m]
movies_filtered
```

```
def weight_rating(df, m=m, C=C):
    R = df['vote_average']
    v = df['vote_count']
    wr = ((v/v+m)* R) + (m/(v+m)*C)
    return wr
```

```
movies_filtered['weighted rating'] = movies_filtered.apply(weight_rating,axis=1)
movies_filtered
```

```
movies_filtered.sort_values('weighted rating', ascending = False)[['title','weighted rating']].head(10).to_dict()
```

# Content-based filtering

```
import pandas 
movies = pandas.read_csv('movies_small.csv', sep=';')
```

```
movies.head(10)
```

```
# In content based we will compare the overview of the current movie against all the other movies and will select the
# most similar movies that have matching overviews or very similar overviews
```

```
from sklearn.feature_extraction.text import TfidfVectorizer  # To compare words in the overview
tfidf = TfidfVectorizer(stop_words = 'english') # Removing stop words


#Replacing the non values if there are any
movies['overview'] = movies['overview'].fillna('')  #nan values are replaced by empty strings

movies['overview']
```

```
tfidf_matrix = tfidf.fit_transform(movies['overview'])
```

```
pandas.DataFrame(tfidf_matrix.toarray(), columns = tfidf.get_feature_names())
```

### similarity matrix

```
from sklearn.metrics.pairwise import linear_kernel

similarity_matrics = linear_kernel(tfidf_matrix, tfidf_matrix)

similarity_matrics
```

### Find the most similar movies to a certain movie

```
movie_title ='John Carter'
index = movies.loc[movies['title'] == movie_title]
index
```

```
scores = list(enumerate(similarity_matrics[index]))
scores
```

```
sorted(scores, key = lambda x:x[1], reverse=True)
```

```
movie_indices = [index[0] for index in scores[1:4]]
movie_indices 
```

```
movies['title'].iloc[movie_indices]
```

```
def similar_movies(movie_title, no_of_movies):

    index = movies.loc[movies['title'] == movie_title]
    scores = list(enumerate(similarity_matrics[index])) #Similarity Scores with other movies
    scores = sorted(scores, key = lambda x:x[1], reverse=True) # Sorting the scores
    movie_indices = [index[0] for index in scores[1:no_of_movies]]
    similar_titles = list(movies['title'].iloc[movie_indices])

    return similar_titles
```

```
print(similar_movies('Cars 2',3))
```
